### Import Paket Pandas dan Membaca Dataset
Agar bisa menggunakan paket pandas kita  perlu melakukan import paket terlebih dahulu, dan kita dapar mengimpor pandas sebagai pd, matplotlib.pyplot sebagai plt dan seaborn sebagai sns agar lebih singkat.Setelah itu  kita deklarasikan DataFrame (df) dengan pd.read_csv untuk membaca file csv agar lebih mudah dan efisien.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("ParisHousingClass.csv")
```

### Grouping Data Berdasarkan Kategori
pada bagian ini saya mengelompokkan data berdasarkan "category"

```python
category = df.groupby('category').sum()
category.head()
```

## Visualisasi Data  
### 1. Grafik Batang dan Diagram Lingkaran(Pie Chart)
Pada bagian ini saya memvisualisasikan data dengan diagram batang dan diagram lingkaran dalam satu cell kode yaitu dengan menggunakan subplot

```python
plt.figure(figsize=(15,5)) #ukuran diagram 

plt.subplot(1,2,1) #subplot 
plt.title('Jumlah Rumah Berdasarkan Kategori') #title / judul diagram
barplot = sns.barplot(x=category.index,y=df['category'].value_counts())
barplot.set(xlabel='kategori', ylabel='total', ylim=(1000,9000)) #label x, y dan limit nilai y

plt.subplot(1,2,2)
plt.title('Persentase Kepemilikan Kolam Renang')
df['hasPool'].value_counts().plot.pie(autopct = '%.2f%%', labels={"0 = Tidak Ada":0,"1 = Ada":1})
```

Dari diagram batang diatas kita dapat mengetahui bahwa total / jumlah rumah di paris dengan kategori "Basic" memiliki jumlah yang jauh lebih tinggi yaitu diatas 8000 unit dibandingkan dengan rumah dengan kategori "Luxury" dengan jumlah dibawah 2000 unit bahkan tidak sampai 1500 unit.

Dari diagram lingkaran diatas kita dapat mengetahui bahwa 50.32% rumah yang ada di paris tidak memiliki kolam renang dan 49.68% sisanya memiliki kolam renang. Artinya hampir setengah dari total rumah yang ada di paris memiliki kolam renang (0 = tidak memiliki kolam renang, 1 = memiliki kolam renang).

### 2. Histogram dan Lineplot

```python
plt.figure(figsize=(15,5)) #ukuran diagram 

plt.subplot(1,2,1) #subplot 
plt.title('Frekuensi Pembangunan Rumah di Paris Pertahun')
plt.xlabel('Tahun')
df['made'].plot.hist(bins=32)

plt.subplot(1,2,2) #subplot 
plt.title('Diagram Lineplot Pembangunan Rumah di Paris Pertahun')
dibuat = df['made'].value_counts()
sns.lineplot(dibuat).set(xlabel='Tahun', ylabel='Frekuensi')
```

Dari histogram diatas kita dapat mengetahui bahwa frekuensi atau banyaknya pembangunan rumah di paris memiliki titik tertinggi yaitu 350 dan titik terendah dibawah 300.

Dari diagram lineplot diatas kita dapat mengetahui bahwa perkembangan pembangunan rumah di paris mengalami lonjakan yang tidak stabil dengan titik tertinggi pada angka diatas 350 dan titik terendah dibawah 290.

# Heatmap

```python
total_corr = df.corr()
plt.subplots(figsize=(20, 20))
plt.title('Korelasi Antar Data dengan Diagram Heatmap')
sns.heatmap(total_corr, annot=True, cmap='turbo')
plt.show()
```

Dari diagram heatmap diatas kita dapat mengetahui bahwa kolom 'price' dan 'squareMeters' saling berkorelasi yaitu semakin tinggi harga rumah maka luas bangunan juga akan semakin besar/luas

```python
plt.title('Diagram Boxplot Ukuran Rumah')
df['squareMeters'].plot(kind='box',figsize=(12,5))
plt.show()
```

Dari diagram boxplot diatas kita dapat mengetahui bahwa median data "squareMeters" sekitar 50000, kuartil bawah sekitar 20000, kuartil atas sekitar 75000. Dengan nilai minimum dibawah 20000(Sekitar 0) dan maksimum 100000.
